# Waiter Tips (Case Study)
The food server of a restaurant recorded data about the tips given to the waiters for serving the food. The data recorded by the food server is as follows:
<br>**total_bill:** Total bill in dollars including taxes
<br>**tip:** Tip given to waiters in dollars
<br>**sex:** gender of the person paying the bill
<br>**smoker:** whether the person smoked or not
<br>**day:** day of the week
<br>**time:** lunch or dinner
<br>**size:** number of people in a table 
    
Based on this data, our task is to find the factors affecting waiter tips.


### Waiter Tips Prediction using Python
We begin the task of waiter tips analysis by importing the necessary Python libraries and the dataset.


```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt

#import dataset
waiter_tips = pd.read_csv("Waiter Tips.csv")

#view first five rows of dataset
print(waiter_tips.head())
```



Now letâ€™s see the **tips given to the waiters according to the days** to find out which day the most tips are given to the waiters:


```python
# if using jupyter notebook include:
%matplotlib inline
#tips given to the waiters according to the days 
waiter_tips.groupby(["day"]).sum().plot(kind="pie", y="tip", autopct="%.1f%%")
plt.show()
```

From the above plot, we can see that **most tips were given to waiters on Saturday.**



Now letâ€™s look at **the number of tips given to waiters by gender** of the person paying the bill to see who tips waiters the most:

```python
# if using jupyter notebook include:
%matplotlib inline
#tips given to the waiters according to the gender
waiter_tips.groupby(["sex"]).sum().plot(kind="pie", y="tip", autopct="%.1f%%")
plt.show()
```

From the above plot, we can see that **most tips were given to waiters by Males.**



Now letâ€™s see if **a smoker tips more or a non-smoker:**

```python
# if using jupyter notebook include:
%matplotlib inline
#Do smokers tip more than non smokers
waiter_tips.groupby(["smoker"]).sum().plot(kind="pie", y="tip", autopct="%.1f%%")
plt.show()
```

From the above plot, we can see that **most tips were given to waiters by non smokers.**



Now letâ€™s see if **most tips are given during lunch or dinner:**

```python
# if using jupyter notebook include:
%matplotlib inline
#tips given to the waiters according to time of meal
waiter_tips.groupby(["time"]).sum().plot(kind="pie", y="tip", autopct="%.1f%%")
plt.show()
```

From the above plot, we can see that **most tips were given to waiters during dinner.**
