```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#import dataset
waiter_tips = pd.read_csv("Waiter Tips.csv")

#view first five rows of dataset
print(waiter_tips.head())
```

```python
# tips given to the waiters according to: the total bill paid, number of people at a table, and the day of the week:
sns.set_style("darkgrid")
sns.relplot(x="total_bill", y="tip", hue="day", style="size", kind="scatter", data=waiter_tips)
plt.show()
```

```python
#tips given to the waiters according to: the total bill paid, the number of people at a table, and the gender of the person paying the bill:
sns.set_style("darkgrid")
sns.relplot(x="total_bill", y="tip", hue="sex", style="size", kind="scatter", data=waiter_tips)
plt.show()
```

```python
#tips given to the waiters according to: the total bill paid, the number of people at a table, and the time of the meal:
sns.set_style("darkgrid")
sns.relplot(x="total_bill", y="tip", hue="time", style="size", kind="scatter", data=waiter_tips)
plt.show()
```

```python
import matplotlib.pyplot as plt
# if using jupyter notebook include:
%matplotlib inline
#tips given to the waiters according to the days 
waiter_tips.groupby(["day"]).sum().plot(kind="pie", y="tip", autopct="%.1f%%")
plt.show()
```

From the above plot, we can see that most tips were given to waiters on Sunday.

```python
# if using jupyter notebook include:
%matplotlib inline
#tips given to the waiters according to the gender
waiter_tips.groupby(["sex"]).sum().plot(kind="pie", y="tip", autopct="%.1f%%")
plt.show()
```

From the above plot, we can see that most tips were given to waiters by Males.

```python
# if using jupyter notebook include:
%matplotlib inline
#Do smokers tip more than non smokers
waiter_tips.groupby(["smoker"]).sum().plot(kind="pie", y="tip", autopct="%.1f%%")
plt.show()
```

From the above plot, we can see that most tips were given to waiters by non smokers.

```python
# if using jupyter notebook include:
%matplotlib inline
#tips given to the waiters according to time of meal
waiter_tips.groupby(["time"]).sum().plot(kind="pie", y="tip", autopct="%.1f%%")
plt.show()
```

From the above plot, we can see that most tips were given to waiters during dinner.
